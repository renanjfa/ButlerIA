### *3. Data Quality Audit*

#### Pandas functions used in the cell below:

_df.shape[0]_ = number of rows

_df.shape[1]_ = number of columns

_df.columns_ = returns an object columns

_list(df.columns)_ = transform into a list of col_names

_df.dtypes_ = returns an object for dataframe datatypes

_df.isnull().sum().sum()_ = returns number of missing values in dataframe

_df.duplicated()_ = returns a mask for duplicated rows

_df[df.duplicated()]_ = returns dataframe filtered with the mask condition

```python
print("=====================")
print("Data Quality Audit")
print("=====================\n")

# dimension
print("\n1. DATASET DIMENSIONS:")
print(f"    Rows: {df.shape[0]}")
print(f"    Columns: {df.shape[1]}\n")

# columns names
print("\n2. COLUMNS NAMES:")

col_names = list(df.columns)

for index, col in enumerate(col_names):
    print(f"    {index}. {col}")

# data types of columns
print("\n3. COLUMN DATA TYPES:")
datatypes = df.dtypes
print(f"{datatypes}")

# missing values
print(f"\n4. MISSING VALUES: {df.isna().sum().sum()}")


# duplicate rows
duplicates_mask = df.duplicated()
print(f"\n5. DUPLICATE ROWS: {len(df[duplicates_mask])}")

# dataframe head
print("\n6. FIRST 5 ROWS:")
df.head()
```
